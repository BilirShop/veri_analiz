```python
# kütüphaneleri ekleyelim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from collections import Counter 

#Python da uyarıları kapatalım
import warnings 
warnings.filterwarnings("ignore")
#veriyi içeri aktaralım
veri = pd.read_excel("YkProfili-07082022191344.xlsx")

veri_date = veri["Profil Tarihi"]
veri_date1=veri_date.str.replace('.2022', '')
veri_date2 = veri_date1.str.slice(stop=-3)
i=0
for x in veri_date2:
    i=i+1
    
    if len(x)<10:
        veri_date2[i-1]="0"+x
veri_date2=veri_date2.str.replace(' ', '-')
#aktif endex işlemleri 
veri_rep=veri["Aktif Endeks"].str.replace(',', '')
veri_rep2=veri_rep.str.replace('.', '')
veri_rep2 = veri_rep2.astype(float)

#endeksler arası fark 15dk da bir yakılan enerji 
for i,x in enumerate(veri_rep2) :
    #print(i)
    if i>0:
        veri_rep[i]=x-veri_rep2[i-1]
    else:
        veri_rep[i]=0
        
# Create figure
fig = go.Figure()

fig.add_trace(
    go.Scatter(x=veri_date2, y=veri_rep))

# Set title
fig.update_layout(
    margin=dict(l=30, r=20, t=40, b=20),
        paper_bgcolor="LightSteelBlue",
    title_text="Time series with range slider and selectors"
)


# Add range slider
fig.update_layout(
    xaxis=dict(
        rangeselector=dict( ),
        rangeslider=dict(
            visible=True
        )
    )
)
        
veri_rep   

#fig = go.Figure(data=go.Scatter(x=veri_date, y=veri_rep))
fig.show()
```
